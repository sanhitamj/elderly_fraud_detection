```python
X = dfc.iloc[:, :-1].values
```

```python
X.shape
```

```python
ones = np.ones((len(X),1))
ones.shape
```

```python
X_intercept = np.column_stack((ones, X))
```

```python
y = dfc.iloc[:, -1]
```

```python
lr = LR()
lr.fit(X_intercept, y)
lr.score(X_intercept, y)
```

```python
model = OLS(y, X_intercept).fit()
summary = model.summary()
summary
```

```python
plt.plot(model.params.values)
plt.xlabel('feature numbers')
plt.ylabel('beta coef')
```
