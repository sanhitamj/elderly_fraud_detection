```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Personal Misc.csv', header=None, 
                 names = ['CasFilID', 'Norm Count', 'Average Expense', 'Neighbours'])
```

```python
df.head()
```

```python
col = (df['Neighbours'] > (df['Neighbours'].max()/10)).replace({True : 'g', False: 'r'})
# .map({True : 'g', False: 'r'})

plt.figure(figsize=(12, 8))
plt.xlabel('Expense normalized to total spending')
plt.ylabel('Number of transactions per day')
# plt.xlim(-100, 10000)
plt.scatter(df['Average Expense'], df['Norm Count'], c=col, alpha=0.2)
```
